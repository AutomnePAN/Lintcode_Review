## Permutation
----------------------------------------------

Given a list of numbers, return all possible permutations.

## Recursion Solution

----------------------------------------------------------------------------------------------------------------

**Time Complexity: $O(n!)$**

```python
class Solution:
    """
    @param: nums: A list of integers.
    @return: A list of permutations.
    """
    def permute(self, nums):
        # write your code here
        if len(nums) <= 1:
            return [nums]
        
        Res = [];
        for i in range(len(nums)):
            nums_pre = nums[:i] + nums[i+1:]
            res_pre = self.permute(nums_pre);
            for perm in res_pre:
                Res.append([nums[i]] + perm);
        return Res;
```

```python
s  = Solution()
```

```python
s.permute([i for i in range(10)])
```

## DFS Solution
----------------------------------------------------------------------------------------------------

```python
class Solution:
    """
    @param nums: A list of Integers.
    @return: A list of permutations.
    """
    def permute(self, nums):
        # write your code here
        def _permute(result, temp, nums):
            if nums == []:
                result += [temp]
            else:
                for i in range(len(nums)):
                    _permute(result, temp + [nums[i]], nums[:i] + nums[i+1:])

        if nums is None:
            return []
        
        if nums is []:
            return [[]]

        result = []
        _permute(result, [], sorted(nums))
        return result
```
